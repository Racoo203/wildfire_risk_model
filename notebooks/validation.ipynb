```python
import pandas as pd
```

```python
spring = pd.read_csv("./data/silver/model_datasets/dataset_clean_spring.csv")
summer = pd.read_csv("./data/silver/model_datasets/dataset_clean_summer.csv")
fall = pd.read_csv("./data/silver/model_datasets/dataset_clean_fall.csv")
```

```python
def show_summary(df):
    temp_df = df.drop(columns=["label"], inplace = False).copy()
    summary_df = temp_df.describe().T
    summary_df["nan (%)"] = 100 * temp_df.isna().sum() / len(temp_df)
    return summary_df
```

```python
show_summary(spring)
```

```python
show_summary(summer)
```

```python
show_summary(fall)
```
